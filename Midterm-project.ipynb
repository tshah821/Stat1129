```python
#Question 1
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as mp


numb = open(r"C:\Users\Plast\OneDrive\Desktop\Mid-term Project\numbers.txt","r")
f =numb.read()
numb1 = f.replace(",","")
numb2 = list(numb1)

def freqs(my_list):
    newdic = {}
    for item in my_list:
        if (item in newdic):
            newdic[item] += 1
        else:
            newdic[item] = 1
    
    return(newdic)

freqs(numb2)

```

```python
#Question 2
amazon = pd.read_csv("C:\\Users\\Plast\\OneDrive\\Desktop\\Mid-term Project\\amazon-orders.csv")
df = pd.DataFrame(data=amazon)
df
```

```python
amazondf = df.loc[:,["Order Date","Tax Charged","Total Charged"]]
amazondf["Total Charged"] = amazondf["Total Charged"].str[1:]
amazondf["Tax Charged"] = amazondf["Tax Charged"].str[1:]
#Erase the $ in front of every number
amazondf["Total Charged"] = pd.to_numeric(amazondf["Total Charged"])
amazondf["Tax Charged"] = pd.to_numeric(amazondf["Tax Charged"])
#Convert string into float
amazondf
```

```python
april = amazondf.loc[0:10,"Total Charged"].sum()
may = amazondf.loc[10:28,"Total Charged"].sum()
june = amazondf.loc[28:34,"Total Charged"].sum()
july = amazondf.loc[34:59,"Total Charged"].sum()
#sum the total charged per order in groups of rows organized by month

monthlyspending = {"April":april,"May":may,"June":june,"July":july}
for k,v in monthlyspending.items():
    print(k,v)
```

```python
mean = np.mean(amazondf["Total Charged"])
median = np.median(amazondf["Total Charged"])
stdvar = np.std(amazondf["Total Charged"])
maximum = np.max(amazondf["Total Charged"])
minimum = np.min(amazondf["Total Charged"])

amazonstats = {"mean":mean,"median":median,"standard deviation":stdvar,"max":maximum,"min":minimum}
for k,v in amazonstats.items():
    print(k,"is",v)
```
