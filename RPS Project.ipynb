```python
import pandas as pd                                                
class RPS:
    
    def __init__(self):
        self.option = {'rock':1,'paper':2,'scissors':3}
        self.p1W = 0
        self.p1L = 0
        self.ties = 0
        self.p2W = 0
        self.p2L = 0

                        
    def wincheck(self,opt1,opt2):
        gamenum = (opt1 - opt2)%3
        if gamenum == 0:
            self.ties +=1
            print("tie!")
        elif gamenum == 1:
            self.p1W +=1
            self.p2L +=1
            print("player 1 wins!")
        elif gamenum == 2:
            self.p2W +=1
            self.p1L +=1
            print("player 2 wins!")
            
    def startup(self):
        
        p1choice = input("player 1 would you like to pick rock, paper, or scissors? \n".lower())
        
        if p1choice not in self.option.keys():
            p1choice = input("player 1 please choose rock, paper or scissors \n".lower())
        
        p2choice = input('player 2 would you like to pick rock, paper, or scissors?'.lower())
        
        if p2choice not in self.option.keys():
            p2choice = input("player 2 please choose rock, paper, or scissors \n".lower())
        
        self.wincheck(self.option[p1choice],self.option[p2choice])
        
    def scoreboard(self):
        summary = {"player 1":[self.p1W,self.p1L,self.ties],"player 2": [self.p2W,self.p2L,self.ties]}
        scoretable = pd.DataFrame(data=summary, index = ["wins","losses","ties"])
        print(scoretable)


if __name__ == "__main__":
    rps = RPS()
    rps.startup()
    
    while True:
        scoreprompt = input("would you like to see the scoreboard? (y/n) \n".lower())
        
        
        if scoreprompt == 'y':
            rps.scoreboard()
        elif scoreprompt =='n':
            exit()
        else:
            continue
        
        contprompt = input("would you like to play again? (y/n) \n".lower())
        if contprompt =='y':
            rps.startup()
        elif contprompt =='n':
            break
        else:
            exit()
            
```
