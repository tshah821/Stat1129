```python
#Question 2
import matplotlib.pyplot as mpt

xarray = np.array([1,6])
yarray = np.array([5,10])

mpt.plot(xarray,yarray)
mpt.show()
```

```python
#Question 5

marks = {'Andy':88,'Amy':66,'James':90,'Jules':55,'Arthur':77}

for k,v in marks.items():
    print(k,":",v)
```

```python
mpt.pie(marks.values(), labels=marks.keys())
mpt.legend(loc="lower left")
mpt.title("Marks")
mpt.show
```

```python
mpt.bar(marks.keys(),marks.values())
mpt.title("marks")
mpt.xlabel("names")
mpt.ylabel("grades")
mpt.show
```
